Below are some helper functions to help get you started. They should look familiar from the lesson!

```python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return (lines, line_img)

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)
```

```python
from collections import deque

prev_left = deque(maxlen=30)
prev_right = deque(maxlen=30)

def intersect_lines(line1, line2): 
    # line = [x1 y1 x2 y2]
    points = np.array_split(line1, 2) + np.array_split(line2, 2)
    x1, y1 = min(points, key = lambda a: a[1])
    x2, y2 = max(points, key = lambda a: a[1])
    return [x1, y1, x2, y2]

def line_params(line):
    x1, y1, x2, y2 = line
    m = (y2 - y1)/(x2 - x1)
    c = (y1 - m * x1)
    return (m, c)

def line_length(line):
    x1, y1, x2, y2 = line
    return (y2 - y1)**2 + (x2 - x1)**2

def seggregate_lines(lines, m_tolerance, c_tolerance, m_limit = 0., c_limit = 0.):
    linemap = {}
    for line in lines:
        for x1, y1, x2, y2 in line:
            m = (y2 - y1)/(x2 - x1)
            c = (y1 - m * x1)
            key = next(((km, kc) for (km, kc) in linemap if abs(km - m) <= m_tolerance and abs(kc - c) <= c_tolerance), None)
            l = [x1, y1, x2, y2]
            if key:
                linemap[key].append(l)
            elif abs(m) > m_limit and x1 != x2:
                linemap[(m, c)] = [l]
    return linemap

def partition_lines(lines, limit):
    result = []
    above = [l for l in lines if line_params(l) > tuple(limit)]
    below = [l for l in lines if line_params(l) <= tuple(limit)]
    if above:
        above_ave = np.average(list(map(line_params, above)), axis=0, weights=list(map(line_length, above)))
        above_weight = sum([line_length(l) for l in above])
        result.append((above_ave, above_weight))
    if below:
        below_ave = np.average(list(map(line_params, below)), axis=0, weights=list(map(line_length, below)))
        below_weight = sum([line_length(l) for l in below])        
        result.append((below_ave, below_weight))
    return result    

def consolidate_lines(lines, maxy):
    m_limit = 0.5
    m_tolerance = 0.2
    c_tolerance = 0.08 * maxy
    linemap = seggregate_lines(lines, m_tolerance, c_tolerance, m_limit)
    positive = [l for (s,c),l in linemap.items() if s > 0]
    negative = [l for (s,c),l in linemap.items() if s < 0]
    result = []
    right = None
    if positive: 
        maxpos = max(positive, key=lambda x: sum(line_length(l) for l in x))
        right_ave = np.average(list(map(line_params, maxpos)), axis=0)
        lines, weights = zip(*partition_lines(maxpos, right_ave))
        right = np.average(lines, axis=0, weights=weights)
    left = None   
    if negative:
        maxneg = max(negative, key=lambda x: sum(line_length(l) for l in x))
        left_ave = np.average(list(map(line_params, maxneg)), axis=0)
        lines, weights = zip(*partition_lines(maxneg, left_ave))
        left = np.average(lines, axis=0, weights=weights)
    return left, right

def smoothen_line(line, lines):
    count = len(lines)
    if line == None or count == 0: return line
    weights = [0.75/count] * count
    ave = np.average([line]+list(lines), axis=0, weights=[0.25]+weights)
    return ave
    
def extend_line(line, y_start, y_end):
    m,c = line
    x_start = (y_start - c)/m if m else x1
    x_end = (y_end - c)/m if m else x2
    return [int(x_start), y_start, int(x_end), y_end]


def image_pipeline(image):
    gray = grayscale(image)
    hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
    maxy, maxx, _ = image.shape
    blurred = gaussian_blur(hsv, 5)
    offx = int(maxx/14)
    marx = offx/4
    offy = int(0.55*maxy)
    vertices = np.array([[(offx, maxy), (maxx - offx, maxy), (maxx/2 + marx, offy), (maxx/2 - marx, offy)]], dtype=np.int32)
    
    edges = canny(blurred, 100, 200)
    masked = region_of_interest(edges, vertices)
    lines, line_image = hough_lines(masked, 2, np.pi/180, 10, 15, 2)
    left, right = consolidate_lines(lines, maxy)
    
    #left, right = smoothen_line(left, prev_left), smoothen_line(right, prev_right)
    #if left != None: prev_left.append(left)
    #if right != None: prev_right.append(right)
    
    final_lines = [[extend_line(line, int(0.60*maxy), maxy)] for line in [left, right] if line is not None]
    filtered_line_image = np.zeros_like(line_image) 
    draw_lines(filtered_line_image, final_lines, color=[255,0,0], thickness=12)    
    weighted = weighted_img(filtered_line_image, image)

    return weighted
    

# TODO: Build your pipeline that will draw lane lines on the test_images
# then save them to the test_images directory.
files = [f for f in os.listdir("test_images") if not f.startswith("op")]
fileno = len(files)

for i, file in enumerate(files):
    image = mpimg.imread(os.path.join('test_images', file))
    #printing out some stats and plotting
    #print('This image is:', type(image), 'with dimesions:', image.shape)

    weighted = image_pipeline(image)
    
    fig = plt.figure()
    plt.imshow(weighted)  #call as plt.imshow(gray, cmap='gray') to show a grayscaled image
    
    res = mpimg.imsave(os.path.join("test_images", "op_"+file), weighted)
```

Pipeline stages:
- Conversion to HSV colour space: HSV seems to render better colour contrast than grayscale especially in the challenge video's shadowy sections
- Canny edge detection: We try to detect edges between a low threshold of 100 and a high threshold of 200. That would mean that any gradient with a value below 100 is rejected and anything with a value above 200 is selected as a strong edge. The intermediary gradients are selected if they are connected to a strong edge
- Hough transform: We perform a hough transform on the image to discover lines from edges. I find that using a slightly higher value for `min_line_length` than `threshold` helps find longer lines and the algorithm is stable against surrounding noise
- Consolidation: 
    - We first group all lines by line parameters (slope, intercept) with some tolerance
    - Find the groups with the largest cumulative length for positive and negative slopes. These two groups represent the best estimates for the left and right lanes
    - Find average line parameters (slope, intercept) for each lane, by averaging individual line parameters weighted by their length
- Extension: For both left and right lanes, we use line parameters to extrapolate to the edges of the region of interest

Failure cases:
- Stretches where lane lines are faded or absent
- Jittery transition of lane predictions across subsequent frames

Possible improvements:
- Smoothen lane predictions by averaging across frames. I am ambivalent about this option as it seems like we are smoothening too early i.e. I think it may be useful to retain reality as is until the decision phase
- Line fitting as opposed to averaging line parameters. This seems to be a popular option among classmates but I am worried this will lose the positive bias I want for longer line segments in the lane
